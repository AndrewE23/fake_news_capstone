# Fake News Detection NLP - Visualisations

**Goal:** To generate a couple extra visualisations for an included presentation component, given in full on 20/07/23. This notebook is otherwise optional for carrying out the central task of creating an NLP classification model.

**Note:** I have decided to load and split my preprocessed dataset, as my wordmaps are going to use the tokenized words as opposed to pre-cleaned vocabulary (i.e. I don't want stopwords in them).

```python
# Libraries
import pandas as pd
import numpy as np

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Libraries Load and merge datasets
test_df = pd.read_csv('data/processedISOT/Test.csv')
train_df = pd.read_csv('data/processedISOT/Train.csv')

joint_df = pd.concat([test_df, train_df], axis=0)
joint_df.info()
```

```python
# Split datasets depending on fake_news value
df_fake = joint_df[joint_df['fake_news'] == 1]
df_real = joint_df[joint_df['fake_news'] == 0]

df_fake.info()
```

```python
fake_text = ' '.join(df_fake['tokens'])
real_text = ' '.join(df_real['tokens'])
```

```python
# Generate Word Maps
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(fake_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Generate Word Maps
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(real_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
