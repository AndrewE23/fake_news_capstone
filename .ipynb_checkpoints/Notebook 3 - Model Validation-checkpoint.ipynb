# Fake News Detection NLP - Model Validation

**Goal:** To make an NLP algorithm that can detect whether an article is fake news with a high degree of accuracy. This third notebook is designed to validate two models that were created in Notebook 2, using the remainder of the ISOT fake news dataset alongside some other samples I will prepare and vectorize.

**Hypothesis:** I am going into this with a decent understanding of how NLP algorithms work, so I expect these models to be better at picking up on differences in writing style rather than actual accuracy. Since the "True" news set is filled _only_ with Reuters articles, I also expect my own trained vectors to perform worse on data outside of the ISOT dataset than Google's own pretrained vectors.

```python
##================#
# Libaries
##================#
import pandas as pd
from pandas.errors import ParserError
import numpy as np
import ast
import regex as re

import matplotlib.pyplot as plt
import seaborn as sns
import pickle

import gensim
import gensim.downloader as api
from gensim.models.keyedvectors import KeyedVectors
from gensim.models import Word2Vec

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
##================#
# Google W2V
##================#
# API download word2vec to save on cloud, download needs
vec_path = api.load("word2vec-google-news-300", return_path=True)
print(vec_path)

##================#
# Import Models
##================#
# Unpickle Model A
model_a = pickle.load(open('models/fakenews_model_1.pkl', 'rb'))

# Unpickle Model B
model_b = pickle.load(open('models/fakenews_model_2.pkl', 'rb'))

##================#
# Load Test ISOT
##================#
test_df = pd.read_csv('data/processedISOT/Test.csv')

# Drop "Unnamed: 0" column, which exists for some reason
test_df = test_df.drop('Unnamed: 0', axis=1)

# Quick test
test_df.head()
```

### Repeat Process on Test Data
Basically, we need to vectorize the remaining ISOT data before we can test the models with it.

```python
# Features: X is defined later
y = test_df['fake_news']
```

```python
##================#
# Vectorizer A
##================#
# Predefined argument 1 - Embedding Size
embed_dim = 512

token_lists = test_df['tokens']
sentences = [ast.literal_eval(tokens) for tokens in token_lists if tokens]
w2v_a = Word2Vec(sentences=sentences, vector_size=embed_dim, min_count=1)

# Predefined argument 2 - Size of Vocabulary
vocab_length = len(w2v_a.wv.key_to_index) + 1
```

```python
# Map actual vocabulary to numbers
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
X = tokenizer.texts_to_sequences(sentences)
word_index = tokenizer.word_index

# Predefined argument 3 - Input Length
maxlen = 840

# Pad (and truncate) for consistent lengths; necessary for a neural network
X = pad_sequences(X, maxlen=maxlen)
```

```python
##================#
# Model A Test
##================#
y_pred = (model_a.predict(X) > 0.5).astype("int")
```

```python
# Print accuracy score
print(f"Accuracy score: {accuracy_score(y, y_pred)}")

# Print precision, recall, and F1 scores
print(f"Precision score: {precision_score(y, y_pred)}")
print(f"Recall score: {recall_score(y, y_pred)}")
print(f"F1 score: {f1_score(y, y_pred)}")
```

```python
# Compute the confusion matrix
cm = confusion_matrix(y, y_pred)

# Print the confusion matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens")
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Fake News - Model A')
plt.show()
```

```python
# Compute ROC
fpr, tpr, thresholds = roc_curve(y, y_pred)
roc_auc = auc(fpr, tpr)
print(roc_auc)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='green', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

The model using my custom vectors has an accuracy rating of 73.5%, which is not quite as good as I expected but still decent given the task at hand; especially since it seems more prone to giving false positives (for being fake news) than false negatives. I like this.

Next up is to test using Google's vectors.

```python
##================#
# Model B Test
##================#
y_pred = (model_b.predict(X) > 0.5).astype("int")
```

```python
# Print accuracy score
print(f"Accuracy score: {accuracy_score(y, y_pred)}")

# Print precision, recall, and F1 scores
print(f"Precision score: {precision_score(y, y_pred)}")
print(f"Recall score: {recall_score(y, y_pred)}")
print(f"F1 score: {f1_score(y, y_pred)}")
```

```python
# Compute the confusion matrix
cm = confusion_matrix(y, y_pred)

# Print the confusion matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Fake News - Model B')
plt.show()
```

```python
# Compute ROC
fpr, tpr, thresholds = roc_curve(y, y_pred)
roc_auc = auc(fpr, tpr)
print(roc_auc)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

The model trained using Google's vectors actually performs worse than my own, which in retrospect makes sense. Google's vectors being pretrained on a much larger dataset means there is more room to misinterpret word patterns and used vocabulary on this particular dataset. However, I must also recognize that there is a chance that my own vectors might have learned some flawed, yet consistent patterns due to picking up vocabulary that may not have been used in Google's vectors.

Either way, the model trained on my own vectors is much more sensitive to fake news within the dataset, which I consider to be a win given my goal of using an NLP model to detect and classify fake news. What I must do now is test it on articles not found in the ISOT fake news dataset.

As expected, the models perform slightly worse with this other dataset. This is sensible because they were fitted with data from a particular variety of sources, while this new data appears to be largely based on articles from other sources. As such, there are a couple reasons why the performance may be worse:
1. The models could be slightly overfitted, although this is unlikely due to my prior tests showing that they perform well (but not too well) on the portion of the ISOT dataset that was not used to train them.
2. The Kaggle dataset could have issues ranging from different subjects/vocabulary that the model was not trained on, to data sourcing or quality issues. 
    - The first of these issues would be remedied by having a model that can either learn from new data as it is tested, or else by training it with articles from a wider variety of sources.
    - The second of these issues would require us to work out a more reliable means of harvesting news data, be it via scrapers or else finding a verified source of real & fake news.
3. The ISOT "True" dataset was sourced entirely from Reuters, and it is likely that their "Fake" dataset was likewise sourced from a consistent group of sites. This could lead to models that perform better on articles that have consistent writing style that may not be present in the Kaggle dataset.
4. The ISOT dataset that was used to train the models may need further preparation, on the off chance I missed anything particularly egregious.


All that being said, it is still worth stating that the models are passable at 61.4% (Model A) and 62.3% (Model B) accuracy even when using articles outside of the ISOT dataset. Moreover, the ROC curves show that the models are somewhat efficient when it comes to giving me an accurate classification for a given article. There is likely room for improvement, especially given that the ISOT-trained models seem to perform better on ISOT-provided test data, but overall I am satisfied with the end result.
